## Utils

```python
NAN = N.nan
def Vector(*args):
    return N.array(args)
def repeatFor(mapFn,**kwargs):
    iteration_data = kwargs.get('iteration_data',[])
    metadata       = kwargs.get('metadata',{}) 
    result         = []
    for i in iteration_data:
        res = mapFn(i,**{'metadata':metadata})
        result.append(res)
        metadata['last_result']=res
    return result
def merge(data,fn):
    length = len(data)
    res = None
    for i,datum in enumerate(data):
#         print(i,datum,length)
        if(i==0):
            res = datum
        if(i+1 < length):
            d1 = data[i+1]
            res = fn(res,d1)
    return res
```

```python
MISSING_VALUES_COLUMNS = ["Age","Cabin","Embarked"]
mostFrequentImp        = SimpleImputer(missing_values=NAN, strategy='most_frequent')
data                   = df[MISSING_VALUES_COLUMNS]
transformedData        = mostFrequentImp.fit_transform(data)
def assignColumns(args,**kwargs):
    i,column = args
    metadata = kwargs.get('metadata',{})
    data = metadata['data']
    df = metadata['df']
    df[column]=data[:,i]
    return df
df = repeatFor(assignColumns,iteration_data=enumerate(MISSING_VALUES_COLUMNS),metadata={
    'data':transformedData,
    'df':df
})[-1]
```

```python
COLUMNS_TO_ENCODED = ['Sex','Cabin','Embarked']
columns_data       = df.loc[:,COLUMNS_TO_ENCODED]

def encodedFn(args,**kwargs):
    i,column = args 
    metadata = kwargs.get('metadata',{})
    df = metadata['df']
    le_s =SK.preprocessing.LabelEncoder()
    data =df[[column]]
    new_data = le_s.fit_transform(data)
    df[column] =  new_data
    return df

df = repeatFor(encodedFn,iteration_data = enumerate(COLUMNS_TO_ENCODED) ,metadata={'df':df})[-1]
```

### Random sampling with replacement

```python
resample      = SK.utils.resample
TOTAL_SAMPLES = 5
factor        = .60
sample_size   =  int(Xy_train.shape[0]*factor) 
Ns            = [sample_size]*TOTAL_SAMPLES
replaces      = [True]*TOTAL_SAMPLES

def clousureFn(args,**kwargs):
    i,n,r    = args
    metadata = kwargs.get('metadata',{})
    X        = metadata.get('X',N.random.rand(100,2))
    data     = resample(X,replace=r,n_samples=n)
    return  data
samples = repeatFor( clousureFn ,iteration_data = zip(range(TOTAL_SAMPLES),Ns,replaces) ,metadata={'X':Xy_train})
```

### Classification Algorithm

```python
class ClassificationAlgorithm(object):
    def __init__(self,**kwargs):
        self.metadata          = kwargs
        self.algorithm         = self.metadata.get('algorithm',None)
        self.defaultHyperPrams = self.metadata.get('hyper_parameters',{})
        self.clf               = self.algorithm(**self.defaultHyperPrams)
    def fit(self,X,y,**kwargs):
        hyperParams = kwargs.get('hyper_parameters',None)
        if not hyperParams:
            return self.clf.fit(X,y)
        else:
            self.clf = self.algorithm(**hyperParams).fit(X,y)
            return self.clf
    def predict(self,X):
        return self.clf.predict(X)
    def predict_proba(self,X):
        return self.clf.predict_proba(X)
c00 = ClassificationAlgorithm(
    algorithm=CalibratedClassifierCV,
    hyper_parameters = {
        'base_estimator':SK.linear_model.Perceptron(alpha=1e-01,max_iter=100,tol=1e-01,shuffle=False,eta0=0.7),
        'method':'sigmoid'
    }
)
c01 = ClassificationAlgorithm(
    algorithm=CalibratedClassifierCV,
    hyper_parameters = {
        'base_estimator':SK.linear_model.Perceptron(alpha=1e-02,max_iter=1000,tol=1e-02,shuffle=True,eta0=1),
        'method':'sigmoid'
    }
)
c02 = ClassificationAlgorithm(
    algorithm=CalibratedClassifierCV,
    hyper_parameters = {
        'base_estimator':SK.linear_model.Perceptron(alpha=1e-03,max_iter=10000,tol=1e-03,shuffle=False,eta0=1),
        'method':'sigmoid'
    }
)
c03 = ClassificationAlgorithm(
    algorithm=CalibratedClassifierCV,
    hyper_parameters = {
        'base_estimator':SK.linear_model.Perceptron(alpha=1e-04,max_iter=100000,tol=1e-04,shuffle=True,eta0=1.5),
        'method':'sigmoid'
    }
)
c04 = ClassificationAlgorithm(
    algorithm=CalibratedClassifierCV,
    hyper_parameters = {
        'base_estimator':SK.linear_model.Perceptron(alpha=1e-05,max_iter=10000,tol=1e-05,shuffle=True,eta0=.5),
        'method':'sigmoid'
    }
)
```

```python
def clousureFn(args,**kwargs):
    i,clfAlgorithm = args
    metadata       = kwargs.get('metadata',{})
    X_test         = metadata.get('X')
    return clfAlgorithm.predict_proba(X_test)
# 
modelsLen =len(models)
iteration_data  = zip(range(modelsLen),models )
predictions     = repeatFor(
    clousureFn,
    iteration_data = iteration_data,
#     metadata = {'X':X_test[:10,:] }
    metadata = {'X':X_test }
)
predictions_vec = Vector(*predictions)
predictions_int  = list(map(lambda xs: N.argmax(xs,axis=1) ,predictions))
# predictions_int
# predictions_vec.values
# predictions_vec = predictions_vec.reshape(modelsLen,2)
# N.argmax(predictions_vec,axis=1)
```

**Hard voting** (also known as majority voting), every individual classifier votes for a class, and the majority wins. In statistical terms, the predicted target label of the ensemble is the mode of individually predicted labels.

```python
#  hv_predict_int = ST.mode(predictions_vec,axis=0).mode[0]
hv_predict_int = N.argmax(ST.mode(predictions_vec,axis=0).mode[0],axis=1)
# hv_predict_int
```

**Soft voting**, every individual classifier provides a probability value that a specific data point belongs to a particular target class. The predictions are weighted by the classifier's importance and summed up. Then the target label with the greatest sum of weighted probabilities wins the vote.

```python
sv_prediction_int = N.argmax(N.mean(predictions_vec,axis=0),axis=1)
# sv_prediction_int
```

```python
from sklearn.metrics import accuracy_score
def clousureFn(args,**kwargs):
    i,y_prediction = args
#     print(i,y_prediction)
    metadata       = kwargs.get('metadata',{})
    y_true         = metadata.get('y_true')
    score = accuracy_score(y_true,y_prediction)
    return score
iteration_data = zip(range(len(predictions_int)), predictions_int )
scores = repeatFor(clousureFn,iteration_data= iteration_data,metadata={'y_true':y_test})
```

```python
x= ["sample-{}".format(i) for i in range(len(predictions_int)) ] + ["Hard-voting","soft-voting"]
hv_score = accuracy_score(y_test,hv_predict_int)
sv_score = accuracy_score(y_test,sv_prediction_int)
_,ax = M.subplots(figsize=(10,6))
ax.set_title("Accuracy scores")
SB.barplot(x=x,y=scores+[hv_score,sv_score],ax=ax)
```
